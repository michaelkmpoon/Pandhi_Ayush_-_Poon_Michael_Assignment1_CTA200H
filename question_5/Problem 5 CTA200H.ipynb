```python
#Importing the required modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#First define a function for f(x)
def f(x):
    return x*(x - 1)

#Then define a function for f'(x)
def fprime(x, delta):
    return (f(x + delta) - f(x))/delta

#For a log-spaced range of values from 10e-14 to 10e-4 and x = 1
delta = np.logspace(-14, -4, base=10)
x = 1

#Plotting f'(x) using this log-space
plt.plot(delta,fprime(x,delta), "o-")
plt.xscale("log")
plt.xlabel("delta")
plt.ylabel("fprime(x) for x = 1")
plt.show()
```

We can see an odd behaviour as delta approaches 0, which makes sense as we can't actually set delta = 0 and this is just an approximation.
